```python
# å°†å„ä¸ªæ•°æ®é›†çš„æ•°æ®æ±‡æ€»
from shutil import copy
from tqdm import tqdm

def merge_images(src_path,des_path):
    if not os.path.exists(des_path):
        os.mkdir(des_path)
    for f in tqdm(os.listdir(src_path)):
        src_file=os.path.join(src_path,f)
        copy(src_file,os.path.join(des_path,f))
        
def merge_sheet(sheet_list):
    result=pd.Series()
    for s in sheet_list:
        bbox_sheet=pd.read_csv(s,header=None)
        result=pd.concat((result,bbox_sheet),ignore_index=True)
    result.columns=['file_name','bboxes']
    return result
```

```python
# åˆå¹¶æ•°æ®é›†
root_dir='/root/BaiduDogs/data_detected_total'
merge_images('/root/BaiduDogs/data_detected_kaggle/',root_dir)
merge_images('/root/BaiduDogs/data_detected_train/',root_dir)
merge_images('/root/BaiduDogs/data_detected_test/',root_dir)
merge_images('/root/BaiduDogs/data_detected_oxford/',root_dir)
```

```python
# åˆå¹¶BBOXè®°å½•
l=['../yolo/kaggle_data_detected_sheet.csv',
  '../yolo/train_data_detected_sheet.csv',
  '../yolo/test_data_detected_sheet.csv',
  '../yolo/oxford_data_detected_sheet.csv']
bbox_sheet=merge_sheet(l)
```
