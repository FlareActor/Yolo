# Yolo V2导出检测结果
对百度🐶比赛测试集、训练集、kaggle猫狗大战、Oxford🐶数据集进行检测，挑选出置信度大于0.8的🐶图片，以便进行模型retrain

```python
from model import Yolo
import pandas as pd
```

```python
yolo=Yolo()
```

```python
# 检测百度测试集图片并保存记录
bboxes=yolo.predict(test_path='/root/BaiduDogs/test_data/test_data/',output_path='/root/BaiduDogs/data_detected_test',
                    targets=[16],draw_box=False) # 只检测🐶，并且不绘制BBOX
pd.Series(bboxes).to_csv('test_data_detected_sheet.csv')
```

```python
# 检测百度训练集图片并保存记录
import os
root_dir='/root/BaiduDogs/train_data/train_0-99/'
train_bboxes=pd.Series()
for f in os.listdir(root_dir):
    f_dir=os.path.join(root_dir,f)
    b=yolo.predict(test_path=f_dir,output_path='/root/BaiduDogs/data_detected_train',targets=[16],draw_box=False)
    train_bboxes=pd.concat((train_bboxes,pd.Series(b)))
pd.Series(train_bboxes).to_csv('train_data_detected_sheet.csv')
```

```python
# 检测kaggle图片并保存记录
kaggle_bboxes=yolo.predict(test_path='/root/DogsVsCats/train/dog/',output_path='/root/BaiduDogs/data_detected_kaggle',targets=[16],draw_box=False)
pd.Series(kaggle_bboxes).to_csv('kaggle_data_detected_sheet.csv')
```

```python
# 检测oxford图片并保存记录
oxford_bboxes=yolo.predict(test_path='/root/BaiduDogs/oxford_images/',output_path='/root/BaiduDogs/data_detected_oxford',targets=[16],draw_box=False)
pd.Series(oxford_bboxes).to_csv('oxford_data_detected_sheet.csv')
```

```python
# 将各个数据集的数据汇总
from shutil import copy
from tqdm import tqdm

def merge_images(src_path,des_path):
    if not os.path.exists(des_path):
        os.mkdir(des_path)
    for f in tqdm(os.listdir(src_path)):
        src_file=os.path.join(src_path,f)
        copy(src_file,os.path.join(des_path,f))
        
def merge_sheet(sheet_list):
    result=pd.Series()
    for s in sheet_list:
        bbox_sheet=pd.read_csv(s,header=None)
        result=pd.concat((result,bbox_sheet),ignore_index=True)
    result.columns=['file_name','bboxes']
    return result
```

```python
# 合并数据集
root_dir='/root/BaiduDogs/data_detected_total'
merge_images('/root/BaiduDogs/data_detected_kaggle/',root_dir)
merge_images('/root/BaiduDogs/data_detected_train/',root_dir)
merge_images('/root/BaiduDogs/data_detected_test/',root_dir)
merge_images('/root/BaiduDogs/data_detected_oxford/',root_dir)
```

```python
# 合并BBOX记录
l=['../yolo/kaggle_data_detected_sheet.csv',
  '../yolo/train_data_detected_sheet.csv',
  '../yolo/test_data_detected_sheet.csv',
  '../yolo/oxford_data_detected_sheet.csv']
bbox_sheet=merge_sheet(l)
```

```python
bbox_sheet.to_csv('../yolo/detected_sheet_total.csv',header=None)
```
